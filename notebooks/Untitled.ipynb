```python
import pandas as pd
df = pd.read_parquet("consolidado_prescriptivo.parquet")
# filas con pedido 0 pero demanda > 0
problematicos = df[(df["pedido"] == 0) & (df["prediccion_ARIMA"] > 0)]
print("Total filas con pedido==0 y prediccion>0:", len(problematicos))
print(problematicos.groupby("tienda").size().sort_values(ascending=False))
```

```python
import pandas as pd

forecast = pd.read_csv("resultados_forecast.csv")   # si lo tienes en otro path, ajusta
multi = pd.read_csv("resultados_multi_periodo.csv")
# normalizar nombres de columna si no lo has hecho
multi = multi.rename(columns={"periodo": "semana"}) if "periodo" in multi.columns else multi

# crear set de pares sku-semana en cada df
set_forecast = set(zip(forecast["sku"], forecast["semana"]))
set_multi = set(zip(multi["sku"], multi["semana"]))

faltantes = sorted(list(set_forecast - set_multi))
print("Combos sku-semana en forecast pero NO en multi:", len(faltantes))
# opcional: ver algunos ejemplos
print(faltantes[:10])
```
